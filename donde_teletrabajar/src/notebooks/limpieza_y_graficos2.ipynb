```python
import matplotlib.pyplot as plt
import pandas as pd
import requests
import json
```

```python
# DF_11 Poblacion activa por comunidades

df_11 = pd.read_csv("../Data/poblacion activa por comunidades.csv", sep= ";")
```

```python
df_11
```

```python
filtro11 = (df_11['Sexo'] == 'Ambos sexos') & (df_11['Edad'] == 'De 16 a 64 años') & (df_11['Periodo'] == '2023T1')
df_11 = df_11[filtro11]
df_11['Comunidades y Ciudades Autónomas'] = df_11['Comunidades y Ciudades Autónomas'].str.split(' ', expand=True, n=1).iloc[:, 1]
df_11.rename(columns={'Total': 'Porcentaje poblacion activa 2023'}, inplace=True)
```

```python
df_11
```

```python
df_11 = df_11.drop(df_11.index[0])
df_11.drop(columns=['Edad', 'Sexo', 'Periodo'], inplace=True)
df_11
```

```python
df_11=df_11.reset_index(drop=True)
df_11= df_11.sort_values(by='Porcentaje poblacion activa 2023', ascending=False)
df_11
```

```python
df_11['Porcentaje poblacion activa 2023'] = df_11['Porcentaje poblacion activa 2023'].str.replace(',', '.')
df_11['Porcentaje poblacion activa 2023'] = df_11['Porcentaje poblacion activa 2023'].astype(float)
```

```python
df_11
```

```python
df_11.to_csv('../Data/poblacion_activa.csv', index=False)
```

```python
# he puesto comentadas las lineas de guardar la foto, para no estropear las imagenes ya guardadas
plt.figure(figsize=(12, 6))  
plt.bar(df_11['Comunidades y Ciudades Autónomas'], df_11['Porcentaje poblacion activa 2023'])
plt.ylim(65)
plt.xlabel('Comunidad Autónoma')
plt.ylabel('porcentaje de poblacion activa')
plt.title('Porcentaje de poblacion activa 2023 por comunidades')
plt.xticks(rotation=90)  
#plt.savefig('../img/poblacion_activa.jpg', bbox_inches='tight')
plt.show()
```

```python
#DF_9 empresas_municipio_actividad BARES

df_9 = pd.read_csv("../Data/empresas_municipio_actividad.csv", sep= ";")
df_9.tail(40)
df_9.info()
```

```python
df_9_filtro.drop(columns=['Actividad principal','Estrato de asalariados','Periodo'], inplace=True)
```

```python
df_9_filtro
```

```python
df_9_filtro['Locales servicios de comidas']=df_9_filtro['Locales servicios de comidas'].astype(int)
```

```python
# df_13 RENTA POR HOGAR
df_13 = pd.read_csv("../Data/renta_por_hogar_comunidad.csv", sep= ";")

filtro13 = (df_13['Comunidades y Ciudades Autónomas'] != 'Total Nacional') & (df_13['Renta anual neta media por hogar'] == 'Renta neta media por hogar') & (df_13['Periodo'] == 2022)
renta = df_13[filtro13]
```

```python
renta.rename(columns={'Total' : 'Renta media por hogar'}, inplace=True)
```

```python
renta['Comunidades y Ciudades Autónomas'] = renta['Comunidades y Ciudades Autónomas'].str.split(' ', expand=True, n=1).iloc[:, 1]
```

```python
renta = renta.drop(columns=['Renta anual neta media por hogar', 'Periodo'])
```

```python
renta.reset_index(drop=True, inplace=True)
renta.sort_values('Renta media por hogar', ascending=False, inplace=True)
```

```python
renta.to_csv("../Data/renta_hogar.csv", index=False)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(renta['Comunidades y Ciudades Autónomas'], renta['Renta media por hogar'])
plt.ylim(22)
plt.xlabel('Comunidades Autónomas')
plt.ylabel('Renta por hogar en miles €')
plt.title('Renta por hogar en  2023 por Comunidad Autónoma')
plt.xticks(rotation=90)
plt.tight_layout()
#plt.savefig('../img/renta_hogar.jpg', bbox_inches='tight')
plt.show()
```
