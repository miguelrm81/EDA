```python
df_1 = pd.read_csv("../Data/bares.csv")
df_1
```

```python
df_2 = pd.read_csv("../Data/poblacion.csv")
df_2
```

```python
merged_df = pd.merge(df_1, df_2, on="Comunidades y Ciudades Autónomas", how="inner")
```

```python
merged_df.info()
```

```python
merged_df
```

```python
merged_df.drop(columns='Habitantes por Local de Comidas')
```

```python
merged_df['Habitantes por Local de Comidas'] = merged_df['Poblacion total'] / merged_df['Locales servicios de comidas'].round(2)
```

```python
merged_df
```

```python
merged_df['Habitantes por Local de Comidas'] = merged_df['Habitantes por Local de Comidas'].round(2)
```

```python
merged_df = merged_df.sort_values(by='Habitantes por Local de Comidas', ascending=True)
```

```python
merged_df
```

```python
plt.figure(figsize=(12, 7))  
bars = plt.bar(merged_df['Comunidades y Ciudades Autónomas'], merged_df['Habitantes por Local de Comidas'])
plt.ylim(110)
plt.xlabel('Comunidades y Ciudades Autónomas')
plt.ylabel('Habitantes por Local de Comidas')
plt.title('Habitantes por Local de Comidas por Comunidad')
plt.xticks(rotation=45, ha="right", fontsize=10)  

# Esta formula de San Google, es para poder ponerle los valores especificos a cada barra del grafico 
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                 textcoords='offset points', ha='center', va='bottom')

plt.tight_layout()
#plt.savefig('../Data/bares_habitante.jpg', bbox_inches='tight')  
plt.show()
```

```python
merged_df_indice = merged_df.drop(columns=['Locales servicios de comidas','Poblacion total','Habitantes por Local de Comidas'])
```

```python
merged_df_indice
```

```python
merged_df_indice = merged_df_indice.iloc[::-1].reset_index(drop=True)
merged_df_indice.index = range(1, 20)
merged_df_indice['Puntuacion_Bares'] = merged_df_indice.index
merged_df_indice.index = range(1, 20)
```

```python
merged_df_indice
```

```python
merged_df_indice
```

```python
merged_df_indice.to_csv('../Data/puntuacion_bares.csv', index=False)
```
