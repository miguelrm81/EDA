```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# OBTENCION DE LA PUNTUACION DE IPC

df_7 = pd.read_csv("poblacion_activa.csv")
df_7
```

```python
df_7_indice = df_7.drop(columns='Porcentaje poblacion activa 2023')
df_7_indice = df_7_indice.iloc[::-1].reset_index(drop=True)
df_7_indice.index = range(1, 20)
df_7_indice['Puntuacion_poblacion_activa'] = df_7_indice.index
df_7_indice.index = range(1, 20)
```

```python
df_7_indice.to_csv('puntuacion_poblacion_activa.csv', index=False)
```
