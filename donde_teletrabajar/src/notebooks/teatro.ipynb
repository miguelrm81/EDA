```python
diccionario = { 'Comunidad' : ['Burgos' , 'Madrid'],
                'Poblacion': [355045, 6750336],
                'Teatros' : [7, 97] }
```

```python
df_3 = pd.DataFrame(diccionario)
```

```python
df_3['Teatro x habitante'] = (df_3['Poblacion'] / df_3['Teatros']).round(2)
```

```python
df_3['Puntuacion Teatro'] = [19, 1]
```

```python
df_3
```

```python
df_3.to_csv("../Data/puntuacion_teatro.csv", index=False)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 6))

# Gráfica para "Teatros"
axes[0].bar(df_3['Comunidad'], df_3['Teatros'], color='green')
axes[0].set_title('Teatros por Comunidad')
axes[0].set_xlabel('Comunidad')
axes[0].set_ylabel('Teatros')

# Gráfica para "Teatro x habitante"
axes[1].bar(df_3['Comunidad'], df_3['Teatro x habitante'], color='blue')
axes[1].set_title('Teatro x habitante por Comunidad')
axes[1].set_xlabel('Comunidad')
axes[1].set_ylabel('Teatro x habitante')
#plt.savefig('../img/teatros_comunidad.jpg', bbox_inches='tight')
plt.tight_layout()
plt.show()
```
