```python
import matplotlib.pyplot as plt
import pandas as pd
import requests
import json
```

```python
# DF_11 Poblacion activa por comunidades

df_11 = pd.read_csv("data/poblacion activa por comunidades.csv", sep= ";")
```

```python
df_11
```

```python
filtro11 = (df_11['Sexo'] == 'Ambos sexos') & (df_11['Edad'] == 'De 16 a 64 años') & (df_11['Periodo'] == '2023T1')
df_11 = df_11[filtro11]
df_11['Comunidades y Ciudades Autónomas'] = df_11['Comunidades y Ciudades Autónomas'].str.split(' ', expand=True, n=1).iloc[:, 1]
df_11.rename(columns={'Total': 'Porcentaje poblacion activa 2023'}, inplace=True)
```

```python
df_11
```

```python
df_11 = df_11.drop(df_11.index[0])
df_11.drop(columns=['Edad', 'Sexo', 'Periodo'], inplace=True)
df_11
```

```python
df_11=df_11.reset_index(drop=True)
df_11= df_11.sort_values(by='Porcentaje poblacion activa 2023', ascending=False)
df_11
```

```python
df_11['Porcentaje poblacion activa 2023'] = df_11['Porcentaje poblacion activa 2023'].str.replace(',', '.')
df_11['Porcentaje poblacion activa 2023'] = df_11['Porcentaje poblacion activa 2023'].astype(float)
```

```python
df_11
```

```python
df_11.to_csv('poblacion_activa.csv', index=False)
```

```python
plt.figure(figsize=(12, 6))  
plt.bar(df_11['Comunidades y Ciudades Autónomas'], df_11['Porcentaje poblacion activa 2023'])
plt.ylim(65)
plt.xlabel('Comunidad Autónoma')
plt.ylabel('porcentaje de poblacion activa')
plt.title('Porcentaje de poblacion activa 2023 por comunidades')
plt.xticks(rotation=90)  
plt.savefig('poblacion_activa.jpg', bbox_inches='tight')
plt.show()
```

```python
#DF_9 empresas_municipio_actividad BARES

df_9 = pd.read_csv("data/empresas_municipio_actividad.csv", sep= ";")
df_9.tail(40)
df_9.info()
```

```python
filtro9 = (df_9['Actividad principal'] == '56 Servicios de comidas y bebidas') & (df_9['Estrato de asalariados'] == 'Total') & (df_9['Periodo'] == 2022)
df_9_filtro = df_9.loc[filtro9]
df_9_filtro.drop(df_9_filtro.index[0], inplace=True)
df_9_filtro
```

```python
df_9_filtro.reset_index(drop=True, inplace=True)
```

```python
df_9_filtro.at[18, 'Total'] = '300'
df_9_filtro.at[17, 'Total'] = '273'
```

```python
df_9_filtro['Total'] = df_9_filtro['Total'].astype(str).str.replace('.', '', regex=False)
df_9_filtro
```

```python
df_9_filtro['Comunidades y Ciudades Autónomas'] = df_9_filtro['Comunidades y Ciudades Autónomas'].str.split(' ', expand=True, n=1).iloc[:, 1]
```

```python
df_9_filtro.rename(columns={'Total' : 'Locales servicios de comidas 2022'}, inplace=True)
```

```python
df_9_filtro.drop(columns=['Actividad principal','Estrato de asalariados','Periodo'], inplace=True)
```

```python
df_9_filtro
```

```python
df_9_filtro['Locales servicios de comidas']=df_9_filtro['Locales servicios de comidas'].astype(int)
```

```python
df_9_filtro = df_9_filtro.sort_values('Locales servicios de comidas', ascending=False)
```

```python
df_9_filtro.to_csv('bares.csv', index=False)
```

```python
plt.bar(df_9_filtro['Comunidades y Ciudades Autónomas'], df_9_filtro['Locales servicios de comidas'])
plt.xlabel('Comunidades Autónomas')
plt.ylabel('Cantidad locales de Servicios de Comidas')
plt.title('Locales de Servicios de Comidas por Comunidad Autónoma')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('bares.jpg', bbox_inches='tight')
plt.show()
```
